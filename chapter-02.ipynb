# Python Infrastructure
### Poetry Packaging & Dependency Management

[Poetry](https://python-poetry.org/) is a dependency manager for python that is similar to JavaScript's [npm](https://www.npmjs.com/) or [yarn](https://yarnpkg.com/). This is a preference over [conda](https://docs.conda.io/en/latest/miniconda.html), which is used in the book we are following along with. I think that poetry has its advantages due to the '.toml' and '.lock' files for a project. It is also conceptually more intuitive with it's 'develop', 'build' and 'publish' approach. Poetry uses [pypi](https://pypi.org/) to add python libraries as dependencies to a project. When you add a new library, the pyproject.toml and poetry.lock file will be automatically updated and resolve any dependency conflicts with other libraries added to the project. 

## Docker
Docker started in 2013 as a set of PAAS (Platform as a service) products that use operating system level virtualization to deliver software in packages called 'containers'. We can think of a docker container as a seperate filesystem that includes an operating system, a python runtime environment, and any additional system or development tools to execute software. Containers can run on local machines with Windows or on a cloud instance with a Linux-based OS. 

### Images and Containers
There are two fundamental concepts that need to be understood when using Docker: an image and a container. A docker image can be conflated with a python class and a container can be thought of as an instance of the class. Docker's definition can be found below:

*Docker images are the basis of containers. An image is an ordered collection of root filesystem changes and corresponding execution parameters for use within a container runtime. An image typically contains a union of layered filesystems stacked on top of each other. An image does not have state and it never changes.*

*A container is a runtime instance of a Docker image. A Docker container consists of a Docker image, an execution environment, and a standard set of instructions.*

## Cloud Instances
The biggest buzz in the past decade has been commonly referred to as "the cloud". The cloud is really just remote hardware that enables deployment/hosting of software via the internet. Here-in lies the "cloud war" and debate between hosting solutions like Amazon Web Services, Microsoft Azure, Google Cloud, IBM Watson, etc.. These companies provide various SAAS (software as a service) solutions alongside their IAAS (infrastructure as a service) solutions. At Lilly, the [domino](https://www.dominodatalab.com/) platform is a SAAS platform made available to simplify deployments to the cloud. 

As you'll be able to see in the graphic below, the number of redundant tools and applications that have been marketed and used to aid software deployment has created a somewhat chaotic ecosystem. Domino allows us to focus on the science and analysis we're performing instead of worrying about the infrastructure. We can easily use our poetry setup locally to share and reproduce code, but Domino will allow us to actually deploy code to run on infrastructure. This is great for model discovery and development.
