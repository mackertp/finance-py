# Python Infrastructure
### Virtual Environments and Dependency Management

The standard package manager in python is called [pip](https://pip.pypa.io/en/stable/), which comes downloaded with your version of python. Pip will allow you to install packages from the [Python Packaging Index](https://pypi.org/) (PyPi) sso that you can use them in any of your scripts, notebooks, programs, or python applications. When downloading and installing packages for your project, it is recommended to use a virtual environment. 

A virtual environment isolates the Python interpreter, libraries and scripts installed into it from the “system” level Python interpretor (and other virtual environments). There are a few options for creating virtual environments in python. [Virtualenv](https://virtualenv.pypa.io/en/latest/) is a good starting point, as it has had a subset of the library integrated into the standard python library, under the *venv* module. 

There is also [Pipenv](https://pypi.org/project/pipenv/), which is available on PyPi. My preference is to use [Poetry](https://python-poetry.org/). Poetry's advantages come from the '.toml' and '.lock' files that are created for a given project. It is also conceptually intuitive with it's 'develop', 'build' and 'publish' approach to managing projects. When you add a new dependecy, the pyproject.toml and poetry.lock file will be automatically updated and will resolve any dependency conflicts inside the project.

[Conda](https://docs.conda.io/projects/conda/en/latest/) is used throughout this book. Conda's main advantage is that it works with any programming language and includes many packages that borrow code written in lower level languages such as C++ and Fortran.

## Docker
Docker started in 2013 as a set of PAAS (Platform as a service) products that use operating system level virtualization to deliver software in packages called 'containers'. We can think of a docker container as a seperate filesystem that includes an operating system, a python runtime environment, and any additional system or development tools to execute software. Containers can run on local machines with Windows or on a cloud instance with a Linux-based OS. 

### Images and Containers
There are two fundamental concepts that need to be understood when using Docker: an image and a container. A docker image can be conflated with a python class and a container can be thought of as an instance of the class. Docker's definition can be found below:

*Docker images are the basis of containers. An image is an ordered collection of root filesystem changes and corresponding execution parameters for use within a container runtime. An image typically contains a union of layered filesystems stacked on top of each other. An image does not have state and it never changes.*

*A container is a runtime instance of a Docker image. A Docker container consists of a Docker image, an execution environment, and a standard set of instructions.*
